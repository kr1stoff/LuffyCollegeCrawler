```python
import requests
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'}
url = 'https://passport.17k.com/ck/user/login'

# post 请求携带
form_data = {
    'loginName': '17346570232',
    'password': 'xlg17346570232'
}
rsp = requests.post(url, headers=headers, data=form_data)
print(rsp.status_code)
print(dict(rsp.cookies))
```

session 保持连接

```python
sess = requests.Session()
url = 'https://passport.17k.com/ck/user/login'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'}

form_data = {
    'loginName': '17346570232',
    'password': 'xlg17346570232'
}

rsp = sess.post(url, headers=headers, data=form_data)
print(rsp.status_code)
print(dict(rsp.cookies))
```

```python
url = 'https://user.17k.com/www/bookshelf/read.html'
rsp2 = sess.get(url, headers=headers)
print(rsp2.content.decode())
```
